```python
import pandas as pd
import numpy as np
import re
import shap
import hashlib

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression


# ----------------------------------------------------------
# 1. Load reviews + clean
# ----------------------------------------------------------
reviews = pd.read_csv("reviews.csv", on_bad_lines='skip', engine='python')
reviews["text"] = reviews["text"].fillna("")

target_asin = "B00CNQ7HAU"   # <--- your chosen book


# ----------------------------------------------------------
# 2. Extract all reviews for this ASIN
# ----------------------------------------------------------
book_reviews = reviews[reviews["asin"] == target_asin].copy()
book_reviews = book_reviews.reset_index(drop=True)

print("Total reviews loaded:", len(book_reviews))


# ----------------------------------------------------------
# 3. Remove any injected fake reviews
#    (detect any 8-character hex patterns like FLAG hashes)
# ----------------------------------------------------------
# Pattern: 8 hex characters (looks like: B5AE0648, 5F0C1A33 etc.)
hash_pattern = r"\b[A-Fa-f0-9]{8}\b"

mask_fake = book_reviews["text"].str.contains(hash_pattern, regex=True)
book_reviews = book_reviews[~mask_fake]

print("After removing injected hash reviews:", len(book_reviews))


# ----------------------------------------------------------
# 4. Label suspicious vs genuine
# ----------------------------------------------------------
superlatives = [
    "amazing", "incredible", "awesome", "perfect", "outstanding",
    "best", "fantastic", "excellent", "love", "loved", "brilliant"
]

def label_review(text, rating):
    t = text.lower()
    # Only 5-star considered for suspicious/genuine
    if rating < 4.5:
        return 0
    if len(t.split()) <= 6:
        return 1
    if any(w in t for w in superlatives):
        return 1
    return 0

book_reviews["label"] = [
    label_review(t, r)
    for t, r in zip(book_reviews["text"], book_reviews["rating"])
]

print("Suspicious count:", (book_reviews["label"] == 1).sum())
print("Genuine count:", (book_reviews["label"] == 0).sum())


# ----------------------------------------------------------
# 5. Guarantee at least 1 suspicious review
# ----------------------------------------------------------
if book_reviews["label"].sum() == 0:
    shortest_idx = book_reviews["text"].str.len().idxmin()
    book_reviews.loc[shortest_idx, "label"] = 1
    print("No suspicious reviews → forced shortest review to suspicious.")


# ----------------------------------------------------------
# 6. Train TF-IDF + Logistic Regression
# ----------------------------------------------------------
texts = book_reviews["text"].tolist()
labels = book_reviews["label"].tolist()

vectorizer = TfidfVectorizer(stop_words='english', min_df=5)
X = vectorizer.fit_transform(texts)

model = LogisticRegression(max_iter=500)
model.fit(X, labels)

print("Model trained successfully.")


# ----------------------------------------------------------
# 7. Compute suspicion scores
# ----------------------------------------------------------
scores = model.predict_proba(X)[:, 1]
book_reviews["score"] = scores

# genuine = low scores
genuine_idx = np.where(book_reviews["score"] < 0.5)[0]

print("Genuine review count for SHAP:", len(genuine_idx))

X_genuine = X[genuine_idx]


# ----------------------------------------------------------
# 8. Run SHAP on the genuine reviews
# ----------------------------------------------------------
explainer = shap.LinearExplainer(model, X, feature_perturbation='interventional')
shap_values = explainer.shap_values(X_genuine)


# ----------------------------------------------------------
# 9. Find top 3 words that REDUCE suspicion
# ----------------------------------------------------------
mean_shap = np.array(shap_values).mean(axis=0)
feature_names = vectorizer.get_feature_names_out()

# 3 most negative SHAP words
indices = np.argsort(mean_shap)[:3]
top_words = [feature_names[i] for i in indices]

print("Top authenticity words:", top_words)


# ----------------------------------------------------------
# 10. Compute FLAG3
# ----------------------------------------------------------
numeric_id = "037"
concat = "".join(top_words) + numeric_id

flag3_hash = hashlib.sha256(concat.encode()).hexdigest()[:10]
flag3 = f"FLAG3{{{flag3_hash}}}"

print("FLAG3 =", flag3)
```
